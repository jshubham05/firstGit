```python
import cv2 as cv
import io
import matplotlib.pyplot as plt
from PIL import Image
import requests
import numpy as np
```

```python
face_cascade=cv.CascadeClassifier('haar_frontal.xml')

```

```python
cam = cv.VideoCapture(0)
```

```python
image = Image.open(io.BytesIO(requests.get("https://www.parkviewservices.org/wp-content/uploads/2011/11/ParkviewServices-BigPhoto-1.jpg").content)).convert("RGB")
plt.imshow(image)
image=np.array(image)
image.shape
```

```python
image=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
```

```python
plt.imshow(image, cmap='gray')
```

```python
faces=face_cascade.detectMultiScale(image,1.3,5)
```

```python
d=len(faces)
d
```

```python
faces[4]
```

```python
x,y,w,z=faces[4]
```

```python
for face in faces:
    draw=cv.rectangle(image,(x,y),(x+w,y+z),(200,100,1),2)
plt.imshow(draw)
```

```python
plt.imshow(draw)
```

```python
while True:
    flag,image=cam.read()
    cv.imshow("face",image)
    if cv.waitKey(1) & 0xFF==ord('q'):
        break
cam.release()
cv.destroyAllWindows()
```
